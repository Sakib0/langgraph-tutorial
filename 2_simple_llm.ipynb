```python
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os
```

```python
load_dotenv()
# Load HUGGINGFACE_API_KEY from environment 
api_key = os.environ.get("HUGGINGFACEHUB_API_TOKEN")
if not api_key:
    raise RuntimeError('HUGGINGFACEHUB_API_TOKEN not set. Add it to .env or set the environment variable.')
```

```python


# 2. Initialize Endpoint with the correct task
llm = HuggingFaceEndpoint(
    repo_id="deepseek-ai/DeepSeek-R1",
    task="text-generation",  
    max_new_tokens=512,
    temperature=0.7,
    do_sample=True,          # explicit sampling usually helps avoid errors
    stop_sequences=["</s>"]  # Good practice to prevent rambling
)

# 3. Wrap in Chat Interface
model = ChatHuggingFace(llm=llm)

# 4. Create the Chain
prompt = PromptTemplate(
    template='Generate 5 interesting facts about {topic}',
    input_variables=['topic']
)
parser = StrOutputParser()
chain = prompt | model | parser

# 5. Invoke
try:
    result = chain.invoke({'topic': 'Football'})
    print(result)
except Exception as e:
    print(f"Error: {e}")
```
